```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
import pickle
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("./credit.csv")
```

```python
data
```

```python
data.describe
```

```python
data["type"] = data["type"].map({ "PAYMENT": 2, 
                                 "CASH_IN": 3, "TRANSFER": 4,
                                 "DEBIT": 5})
data["isFraud"] = data["isFraud"].map({0: "No Fraud", 1: "Fraud"})
data = data.dropna()
```

```python
x = np.array(data[["type", "amount", "oldbalanceOrg", "newbalanceOrig"]])
y = np.array(data[["isFraud"]])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state=42)
```

```python
x_test
```

```python
model1 = DecisionTreeClassifier()
model1.fit(x_train, y_train)
```

```python
model2 = AdaBoostClassifier(estimator=DecisionTreeClassifier(), n_estimators=50, learning_rate=0.5)
model2.fit(x_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier  
model3= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
model3.fit(x_train, y_train)
```

```python
y_pred1 = model1.predict(x_test)
y_pred2 = model2.predict(x_test)
y_pred3 = model3.predict(x_test)
```

```python
print("Decision Tree:",accuracy_score(y_pred1, y_test))
print("Adaptive Boost Classifier:", accuracy_score(y_pred2, y_test))
print("K Nearest Neighbours:", accuracy_score(y_pred2, y_test))
```

```python
pickle.dump(model1, open('DecisionTreeModel.pkl','wb'))
pickle.dump(model2, open('AdaptiveBoostModel.pkl','wb'))
pickle.dump(model3, open('KNNModel.pkl','wb'))
```
